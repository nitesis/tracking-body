```python
from ultralytics import YOLO
import cv2
```

```python
# Load the pose estimation model
model = YOLO("yolo11n-pose.pt")  # Use the lightweight model for real-time performance

# Open webcam feed
cap = cv2.VideoCapture(2)  # 0 is the default webcam; change if using an external camera
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform pose estimation
    results = model.predict(source=frame, show=False, conf=0.5)  # Perform pose estimation with confidence threshold
    
    # Visualize results
    annotated_frame = results[0].plot()  # Draw detected poses on the frame

    # Display the frame
    cv2.imshow("Pose Estimation", annotated_frame)

    # Exit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
